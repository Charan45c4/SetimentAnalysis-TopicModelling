```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
df=pd.read_csv("BritishAirlinesData.csv")
lemmatizer=WordNetLemmatizer()
stop_words=set(stopwords.words('english'))
df
```

```python
df['cleaned']=df['reviews'].apply(
    lambda text:' '.join([
        lemmatizer.lemmatize(word) 
        for word in re.sub(r'[^a-z\s]','',str(text).lower()).split() 
        if word not in stop_words
    ])
)
df
```

```python
df.dropna(subset=['cleaned'], inplace=True)
df.head()
```

```python
df.to_csv('CleanedFile.csv', index=False)
```
